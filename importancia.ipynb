```python
import matplotlib.pyplot as plt
from matplotlib.cm import get_cmap

# Definição dos dados para o gráfico de pizza
aspectos = [
    'Tomada de Decisão Estratégica',
    'Otimização de Processos e Eficiência',
    'Previsão de Tendências de Mercado',
    'Personalização da Experiência do Cliente',
    'Obtenção de Vantagem Competitiva'
]
importancia = [30, 25, 20, 15, 10]  # Valores em porcentagem

# Gerar cores usando um mapa de cores (Viridis)
cmap = get_cmap('viridis')
cores = cmap([i / len(aspectos) for i in range(len(aspectos))])

# Estilo do gráfico
plt.style.use('seaborn-v0_8-whitegrid')

# Criar figura com tamanho personalizado
plt.figure(figsize=(12, 8))

# Criar o gráfico de pizza
plt.pie(importancia, 
        labels=aspectos,
        colors=cores,
        autopct='%1.1f%%',  # Formato da porcentagem
        startangle=140,      # Ângulo inicial para melhor visualização
        shadow=True,         # Sombra para efeito 3D
        explode=(0.1, 0, 0, 0, 0)  # Destacar a primeira fatia
       )

# Adicionar título com mais destaque
plt.title('A Importância Estratégica da Área de Dados para Empresas', pad=20, fontsize=16, weight='bold')

# Adicionar uma legenda informativa
plt.legend(aspectos, 
           title="Aspectos-Chave",
           loc="center left",
           bbox_to_anchor=(1, 0, 0.5, 1))

# Garantir que o gráfico seja perfeitamente circular
plt.axis('equal')

# Ajustar o layout para evitar sobreposições
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
